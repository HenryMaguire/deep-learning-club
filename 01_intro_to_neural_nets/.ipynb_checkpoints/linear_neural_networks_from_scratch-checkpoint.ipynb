```python
import matplotlib.pyplot as plt
import numpy as np
import pathlib
import time

DATASETS_PATH = pathlib.Path().resolve() / 'datasets'
```

```python

# Setting the random seed, feel free to change it and see different solutions.
np.random.seed(42)

def stepFunction(t):
    if t >= 0:
        return 1
    return 0

def prediction(X, W, b):
    # can be 1 if x>00, or 0 if x<0 
    return stepFunction((np.matmul(X,W)+b)[0])

# Fill in the code below to implement the perceptron trick.
# The function should receive as inputs the data X, the labels y,
# the weights W (as an array), and the bias b,
# update the weights and bias W, b, according to the perceptron algorithm,
# and return W and b.
def perceptronStep(X, y, W, b, learn_rate = 0.01):
    for i in range(len(X)):
        # For each input, find out what the prediction is
        y_hat = prediction(X[i], W, b)
        # if the prediction 
        if (y[i] - y_hat) == 1:
            W[0] += X[i][0]*learn_rate
            W[1] += X[i][1]*learn_rate
            b += learn_rate
        elif (y[i] - y_hat) == -1:
            W[0] -= X[i][0]*learn_rate
            W[1] -= X[i][1]*learn_rate
            b -= learn_rate
    return W, b
    
# This function runs the perceptron algorithm repeatedly on the dataset,
# and returns a few of the boundary lines obtained in the iterations,
# for plotting purposes.
# Feel free to play with the learning rate and the num_epochs,
# and see your results plotted below.
def trainPerceptronAlgorithm(X, y, learn_rate = 0.01, num_epochs = 25):
    x_min, x_max = min(X.T[0]), max(X.T[0])
    y_min, y_max = min(X.T[1]), max(X.T[1])
    W = np.array(np.random.rand(2,1))
    b = np.random.rand(1)[0] + x_max
    # These are the solution lines that get plotted below.
    boundary_lines = []
    for i in range(num_epochs):
        # In each epoch, we apply the perceptron step.
        W, b = perceptronStep(X, y, W, b, learn_rate)
        boundary_lines.append((-W[0]/W[1], -b/W[1]))
    return boundary_lines


```

```python
import csv
X = []
y = []
with open(DATASETS_PATH / "l1_perceptron_data.csv") as f:
    spamreader = csv.reader(f, delimiter=',')
    for row in spamreader:
        X.append([float(s) for s in row[0:2]])
        y.append(float(row[2]))
X = np.array(X)
y = np.array(y)
```

```python
boundary_lines = trainPerceptronAlgorithm(X, y)
```

```python
# step 3: 
def build_W(X_train_, scale=0.1):
    num_features = X_train_.shape[1]
    return np.random.normal(size=(n_classes, num_features),
                            scale=scale)


W  = build_W(X)
```

```python


def L_i(x, y, W):
    """
    unvectorized version. Compute the multiclass svm loss for a single example (x,y)
    - x is a column vector representing an image (e.g. 3073 x 1 in CIFAR-10)
    with an appended bias dimension in the 3073-rd position (i.e. bias trick)
    - y is an integer giving index of correct class (e.g. between 0 and 9 in CIFAR-10)
    - W is the weight matrix (e.g. 10 x 3073 in CIFAR-10)
    """
    delta = 1.0 # see notes about delta later in this section
    
    scores = W.dot(x) # scores becomes of size 10 x 1, the scores for each class
    correct_class_score = scores[y]
    D = W.shape[0] # number of classes, e.g. 10
    loss_i = 0.0
    for j in range(D): # iterate over all wrong classes
        if j == y:
            # skip for the true class to only loop over incorrect classes
            continue
        # accumulate loss for the i-th example
        loss_i += max(0, scores[j] - correct_class_score + delta)
    return loss_i


ti = time.time()
L_ = np.zeros(shape=(X.shape[0],1))


for i, (x, y) in enumerate(zip(X, y_train)):
    L_[i] = L_i(x, y, W)
    
L_for = np.array(L_).reshape(-1, 1)

print(f"Took {time.time() - ti} seconds")
```

Since the weights are all small due to initialisation, we expect the mean loss to be `num_classes - 1` because each incorrect class picks up a 1 (from the delta margin) and the correct class doesn't contribute.

```python
def L_i_vectorized(x, y, W):
    """
    A faster half-vectorized implementation. half-vectorized
    refers to the fact that for a single example the implementation contains
    no for loops, but there is still one loop over the examples (outside this function)
    """
    delta = 1.0
    scores = W.dot(x)
    # compute the margins for all classes in one vector operation
    margins = np.maximum(0, scores - scores[y] + delta)
    # on y-th position scores[y] - scores[y] canceled and gave delta. We want
    # to ignore the y-th position and only consider margin on max wrong class
    margins[y] = 0
    loss_i = np.sum(margins)
    return loss_i

ti = time.time()
L_ = np.zeros(shape=(X.shape[0],1))

for i, (x, y) in enumerate(zip(X, y_train)):
    L_[i] = L_i_vectorized(x, y, W)

print_summary(W)
print_summary(L_)

print(f"Took {time.time() - ti} seconds")
```

```python
def L(X, y, W, sum_all=True, reg=0.0001):
    """
    fully-vectorized implementation :
    - X holds all the training examples as columns (e.g. 3073 x 50,000 in CIFAR-10)
    - y is array of integers specifying correct class (e.g. 50,000-D array)
    - W are weights (e.g. 10 x 3073)
    """
    num_train = X.shape[1]
    num_classes = W.shape[0]
    num_features = W.shape[1]
    #X = X.reshape(X.shape[1], X.shape[0])
    #X = X.T
    delta = 1.0
    # evaluate loss over all examples in X without using any for loops
    # scores will be a 50,000 x 10 matrix
    scores = np.dot(X, W.T) # same as np.dot(W, X.T).T
    
    correct_class_score = np.take_along_axis(scores, y, axis=1)
    
    margins = np.maximum(0, scores - correct_class_score + 1)
    # got to set indices of y to 0
    np.put_along_axis(margins, y, 0, axis=1)
    if sum_all:
        loss = np.mean(margins)
    else:
        loss = np.sum(margins, axis=1)
    
    
    loss += reg * np.sum(W * W)
    
    binary = margins
    binary[margins > 0] = 1
    row_sum = np.sum(binary, axis=1)
    
    binary[np.arange(num_train), y] = -row_sum.T.reshape(-1,1)
    dW = np.dot(X.T, binary)

    # Average
    dW /= num_train
    #print(W.shape, dW.shape)
    # Regularize
    dW += 2 * reg * W.T
    
    return loss, dW

ti = time.time()
loss, dW = L(X, y_train, W, sum_all = False)
print(f"Took {time.time() - ti} seconds")
```

```python
def eval_numerical_gradient(f, x):
    """
    a naive implementation of numerical gradient of f at x
    - f should be a function that takes a single argument
    - x is the point (numpy array) to evaluate the gradient at
    """

    fx = f(x) # evaluate function value at original point
    grad = np.zeros(x.shape)
    h = 0.00001
    #print(grad.shape)
    # iterate over all indexes in x
    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    while not it.finished:
        
        # evaluate function at x+h
        ix = it.multi_index
        old_value = x[ix]
        x[ix] = old_value + h # increment by h
        fxh = f(x) # evalute f(x + h)
        x[ix] = old_value # restore to previous value (very important!)
        
        # compute the partial derivative
        grad[ix] = (fxh - fx) / h # the slope
        if ix[1] % 100 ==0:
            print(ix)
        it.iternext() # step to next dimension
    return grad
```

```python
# to use the generic code above we want a function that takes a single argument
# (the weights in our case) so we close over X_train and Y_train
def CIFAR10_loss_fun(W):
    return L(X, y_train, W)

W = np.random.rand(10, 3073) * 0.001 # random weight vector
df = eval_numerical_gradient(CIFAR10_loss_fun, W) # get the gradient
```

```python
# Vanilla Gradient Descent
losses = []
W_ = copy.copy(W)
step_size = 0.001
no_improvement = 0
while True:
    #weights_grad = evaluate_gradient(loss_fun, data, weights)
    loss, dW = L(X, y_train, W_, sum_all = True)
    W_ +=  -step_size * dW.T # perform parameter update
    if len(losses) > 0:
        if loss > losses[-1]*1.1:
            step_size /= 2
            print(f"dropping step size : {step_size}")
            if step_size <1e-6:
                break
        elif loss > losses[-1]:
            no_improvement += 1
            if no_improvement > 4:
                step_size /= 2
                print(f"dropping step size : {step_size}")
                if step_size <1e-6:
                    break
    losses.append(loss)
    print(loss)
```
